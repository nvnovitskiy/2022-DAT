Работа с пропущенными значениями. Посмотрим, есть ли у нас пропущенные значения.

```python
df.isna().sum()
```

```python
# Генерируем пустые значения в колонке Price_euros
for _ in range(0, 25):
    df.iloc[np.random.randint(11, df.shape[0]), 11] = np.NaN
```

```python
df.isna().sum() # Пустые значения были добавлены
```

```python
df["Price_euros"] = df["Price_euros"].fillna(value=df["Price_euros"].mean())
```

```python
df.isna().sum() # Видим, что от пустых значений мы избавились
```

Видим, что больше пропущенных значений в этой колонке нет, каждый игрок теперь имеет значение по данному параметру

```python
# создание нового поля вычисленного на основе значений других полей
# 1-й способ - через выражение на базе имеющихся колонок
df["Price_rub"] = df["Price_euros"] * 85.89
```

```python
# Выведем 5 случайных строк и посмотрим на результат
df.sample(5)
```

```python
# 2-й способ DataFrame.apply
df["Status"] = df.apply(lambda x: "expensive" if x["Price_rub"] > 100000 else "not expensive", axis=1)
```

```python
# Смотрим результат
df.sample(5)
```

```python
# 3-й способ Series.apply
df["Weight_status"] = df["Weight"].apply(lambda x: "heavy" if x > "2.00kg" else "not heavy")
```

```python
# Смотрим результат
df.sample(5)
```

```python
# Сортировка
df = df.sort_values(by="Price_rub", ascending=False)
df.head(10)
```
